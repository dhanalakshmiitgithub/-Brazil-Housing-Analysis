# Leval 1 Analysis(Univariate Analysis)

```python
def info_categorical(col):
    print(f"Unique values in {col} are: {df[col].unique()}")
    print(f"Mode of {col} is {df[col].mode()}")
    print(f"Number of missing values in {col} is {df[col].isnull().sum()}")
info_categorical
```

# Analysis of City

```python
info_categorical('city')
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
ax[0].set_title("Count of City")                  
percentage = df["city"].value_counts()         
labels = list(df["city"].value_counts().index) 
sb.countplot(x = df["city"], ax = ax[0])
plt.pie(percentage,labels = labels, autopct= "%0.2f%%")  
plt.show()
```

# Analysis of Animal

```python
info_categorical("animal")
```

# Level 1 Analysis of Numerical Columns

```python
def info_numerical(col):
    print(f"The mean of the {col} is {df[col].mean()}")
    print(f"The median of the {col} is {df[col].median()}")
    print(f"The mode of the {col} is {df[col].mode()[0]}")
    print(f"The standard deviation of the {col} is {df[col].std()}")
    print(f"Number of missing values in the {col} is {df[col].isnull().sum()}")
```

1. Analysis of the Area

```python
info_numerical("area")
```

```python
fig, ax=plt.subplots(1,2, figsize=(20,7))
sb.histplot(x=df["area"], ax=ax[1], color='red', edgecolor='black',linewidth=2, kde=True)
sb.boxplot(x=df["area"], ax=ax[0], color='purple')
plt.show()
```
